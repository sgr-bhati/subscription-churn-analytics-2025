```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load Sample
df = pd.read_csv('data/raw/sample_customer_data.csv')
df.head()
```

```python
df.info()
df.describe(include='all')
df.isnull().sum()
```

```python
#Simple Plot
# Count occurrences of each category
counts = df['subscription_canceled'].value_counts()

# Create a bar chart
plt.bar(counts.index, counts.values)
plt.title('Churn distribution (sample)')
plt.show()
```
